```python
import os
import numpy as np
import matplotlib.pyplot as plt
from twaidata.MRI_preprep.io import load_nii_img
```

```python
imgs = []
for imgname in os.listdir(datadir):
    if ".nii.gz" in imgname:
        img, _ = load_nii_img(os.path.join(datadir, imgname))
        imgs.append((imgname, img))
```

```python
islice = 20
for iname, i in imgs:
    print(f"{iname}: {i.shape}")
    plt.imshow(i[:,:,islice])
    plt.show()
```
